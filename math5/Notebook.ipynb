Пусть: 
- $A$ - событие, что клиент привел новых клиентов
- $B$ - событие, что клиент испытал wow-effect
- $C$ - событие, что клиент периодически оставляет нам положительные отзывы
- $D$ - событие, что мы не знаем клиента

Тогда

$P(\text{«клиент приведет нового клиента»}|\text{«клиент испытал wow-effect»})  = P(A|B) = 0.7$ - верное утверждение

2) Если вы хотите найти вероятность, что случайный клиент вашего сервиса приведет нового клиента, что вы возьмёте за разбиение?

$P(A) = \sum\limits_{i=\{B,C,D\}}{P(A|i)P(i)} = 0.02\times0.7+0.2\times0.1+0.78\times0.04 = 0.0652$

Редкое заболевание встречается у $1$% людей. Тест ошибочно показывает наличие болезни у здоровых людей с вероятностью $2$% и ошибочно показывает отсутствие болезни у больных с вероятностью $3$%. Найдите вероятность, что человек на самом деле болен, если тест отрицателен на болезнь.
